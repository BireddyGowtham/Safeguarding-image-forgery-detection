# Exploration of CASIA Datasets

This notebook is used for exploratory data analysis and visualization of the CASIA datasets used for image forgery detection.

```python
import os
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
from torchvision import transforms
from src.dataset import CASIADataset

# Set the paths for the datasets
casia1_path = './data/CASIA1'
casia2_path = './data/CASIA2'

# Load datasets
casia1_dataset = CASIADataset(casia1_path)
casia2_dataset = CASIADataset(casia2_path)

# Display some sample images from CASIA v1
def display_samples(dataset, num_samples=5):
    plt.figure(figsize=(15, 5))
    for i in range(num_samples):
        img, label = dataset[i]
        plt.subplot(1, num_samples, i + 1)
        plt.imshow(img.permute(1, 2, 0).numpy())
        plt.title('Tampered' if label == 1 else 'Authentic')
        plt.axis('off')
    plt.show()

print("CASIA v1 Samples:")
display_samples(casia1_dataset)

print("CASIA v2 Samples:")
display_samples(casia2_dataset)

# Analyze the distribution of classes
def analyze_class_distribution(dataset):
    authentic_count = sum(1 for _, label in dataset.samples if label == 0)
    tampered_count = sum(1 for _, label in dataset.samples if label == 1)
    return authentic_count, tampered_count

authentic_count_v1, tampered_count_v1 = analyze_class_distribution(casia1_dataset)
authentic_count_v2, tampered_count_v2 = analyze_class_distribution(casia2_dataset)

# Plot class distribution
labels = ['Authentic', 'Tampered']
casia1_counts = [authentic_count_v1, tampered_count_v1]
casia2_counts = [authentic_count_v2, tampered_count_v2]

x = np.arange(len(labels))  
width = 0.35  

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, casia1_counts, width, label='CASIA v1')
rects2 = ax.bar(x + width/2, casia2_counts, width, label='CASIA v2')

ax.set_ylabel('Number of Images')
ax.set_title('Class Distribution by Dataset')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

plt.show()
```
